### RAG Pipeline - Data ingestion to Vector DB pipeline

```python
import os
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from pathlib import Path
```

```python
# This function reads all the PDFs in the directory and returns a list of documents

def process_all_pdfs(pdf_directory):
    # Initialize an empty list to hold all documents
    all_documents = []

    # Convert the directory path (String) to a Path object
    pdf_directory = Path(pdf_directory)

    # Find all pdfs recursively (rglob: recursive global search) in the given directory
    # rglob() returns a iterator, Path object of each pdf file found, converts it to a list
    pdf_files = list(pdf_directory.rglob("*.pdf"))

    # Print the number of PDF files found
    print(f"Found {len(pdf_files)} PDF files to process.")

    # Process each PDF file
    for pdf_file in pdf_files:

        # Print the name of the current PDF file being processed
        print(f"Processing {pdf_file.name}")

        try:
            # pdf_file is a Path object of current pdf file, loader expects a string value.
            loader = PyPDFLoader(str(pdf_file))
            # Loads the PDF file and document structure is created (metadata, page content, etc).
            documents = loader.load()

            # Adding more information to the metadata (present inside documents variable).
            for doc in documents:
                doc.metadata["source"] = str(pdf_file)
                doc.metadata["file_type"] = "pdf"

            # We're using extend here instead of append because:
            # Let say doc1 = [1, 2, 3] and doc2 = [4, 5, 6]
            # If we do doc1.append(doc2), we will get [1, 2, 3, [4, 5, 6]] ✕
            # If we do doc1.extend(doc2), we will get [1, 2, 3, 4, 5, 6] ✓
            all_documents.extend(documents)

            # Confirmation message that the PDF file has been loaded.
            print(f"✓ Loaded {len(documents)} pages")

        # If encountered any error while loading the PDF file or a corrupted PDF file, print an error message.
        except Exception as e:
            print(f"✗ Failed to load: {e}")
            
    print(f"Total pages loaded: {len(all_documents)}")
    return all_documents
    
all_pdf_documents = process_all_pdfs("../data")
```
